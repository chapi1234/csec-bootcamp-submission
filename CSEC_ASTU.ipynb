## 2. Given the list [10, 15, 20, 25, 30], write a function that returns the square of only even numbers.


````python
def square_even_numbers(nums):
    return [x**2 for x in nums if x % 2 == 0]

numbers = [10, 15, 20, 25, 30]
square_even_numbers(numbers)
````

## 6. Load the CSV from Q5 and find:  
- The month with the highest total passengers  
- The month with the lowest in 1958


````python
import pandas as pd
from google.colab import files

uploaded = files.upload()
air_df = pd.read_csv("airtravel.csv")
air_df.columns = air_df.columns.str.strip().str.replace('"', '')
print(air_df.columns.tolist())
print(air_df.head())
air_df["Total"] = air_df.iloc[:, 1:].sum(axis=1)
max_month = air_df.loc[air_df["Total"].idxmax(), "Month"]
min_1958_month = air_df.loc[air_df["1958"].idxmin(), "Month"]

print("Month with highest total passengers:", max_month)
print("Month with lowest passengers in 1958:", min_1958_month)
````

## 7. What does the `.groupby()` function do in pandas?  
It splits data into groups based on some criteria, allowing for aggregation and analysis.

### Example:


````python
data = pd.DataFrame({
    "Department": ["IT", "HR", "IT", "HR", "Finance"],
    "Salary": [4000, 3000, 4500, 3500, 5000]
})

data.groupby("Department").mean(numeric_only=True)
````

## 10. What does it mean if a column has a high standard deviation?

It means that the values are spread out far from the mean (more variability).

### Example:


````python
low_std = pd.Series([50, 51, 49, 50, 50])
high_std = pd.Series([10, 90, 50, 30, 100])

print("Low Std Dev:", low_std.std())
print("High Std Dev:", high_std.std())
````
